```python
data = data.map(lambda x,y: (x/255.0,y))
```

```python
# SPLIT DATA
print(len(data))
train_size = int(len(data)*.7)-1
val_size = int(len(data)*.2)+1
test_size = int(len(data)*.1)+1
```

```python
print(f'trina size:{train_size}, val size : {val_size}, Test size : {test_size}')
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

            Building MODEL         

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
# model = Sequential([Conv2D(),...]) #can be done like this also
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3),1,activation = 'relu',input_shape=(256,256,3))) # with 16 filters of size 3,3 and stride of 1
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3),1,activation = 'relu')) # with 16 filters of size 3,3 and stride of 1
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3),1,activation = 'relu')) # with 16 filters of size 3,3 and stride of 1
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('adam',loss = tf.losses.BinaryCrossentropy(),metrics = ['accuracy'])
```

```python
model.summary()
```

TRAIN MODEL

```python
logdir = 'log'
```

```python
# To save model at particular check point
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir = logdir)
```

```python
hist = model.fit(train,epochs=20,validation_data = val, callbacks = [tensorboard_callback]) # hist is history
```

```python
hist.history
```

PLOTING  PERFORMANCE

```python
fig = plt.figure()
plt.plot(hist.history['loss'],color ='teal',label = 'loss')
plt.plot(hist.history['val_loss'],color = 'orange',label = 'val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc = "upper right")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'],color ='teal',label = 'loss')
plt.plot(hist.history['val_accuracy'],color = 'orange',label = 'val_loss')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc = "upper right")
plt.show()
```

EVALUATE PERFORMANCE

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    X,y = batch
    print(len(X))
    yhat = model.predict(X)
    pre.update_state(y,yhat)
    re.update_state(y,yhat)
    acc.update_state(y,yhat)
```

```python
print(f'PRECISION:{pre.result().numpy()}, RECALL:{re.result().numpy()}, ACCURACY:{acc.result().numpy()}')
```
